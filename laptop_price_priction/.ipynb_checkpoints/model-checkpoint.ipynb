```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding='latin-1')
df = df.drop(['Product'], axis=1)
df
```

```python
df = df.join(pd.get_dummies(df.Company))
df = df.join(pd.get_dummies(df.TypeName))
```

```python
df["ScreenResolution"] = df.ScreenResolution.str.split(" ").apply(lambda x: x[-1])
df["ScreenWidth"] =df.ScreenResolution.str.split("x").apply(lambda x: x[0])
df["ScreenHeight"] =df.ScreenResolution.str.split("x").apply(lambda x: x[1])

df = df.drop(["ScreenResolution", "TypeName", "Company",], axis=1)
```

```python
df["CpuBrand"] = df.Cpu.str.split(" ").apply(lambda x: x[0])
df["CpuFrequency"] = df.Cpu.str.split(" ").apply(lambda x: x[-1])
df = df.drop("Cpu", axis=1)
```

```python
df["CpuFrequency"] =df["CpuFrequency"].str[:-3]
df["Ram"] = df["Ram"].str[:-2]
df["Weight"] = df["Weight"].str[:-2]
df
```

```python
df["Ram"] = df["Ram"].astype(int)
df["CpuFrequency"] = df["CpuFrequency"].astype(float)
df["ScreenHeight"] = df["ScreenHeight"].astype(int)
df["ScreenWidth"] = df["ScreenWidth"].astype(int)
```

```python
df
```

```python
df["MemoryAmount"] = df.Memory.str.split(" ").apply(lambda x: x[0])
df["MemoryType"] = df.Memory.str.split(" ").apply(lambda x: x[1])
df
```

```python
def turn_memory_to_MBs(val):
    if "GB" in val:
        return float(val[:val.find("GB")]) * 1000
    elif "TB" in val:
         return float(val[:val.find("TB")]) * 1000000
        
```

```python
df["MemoryAmount"] = df["MemoryAmount"].apply(turn_memory_to_MBs)
```

```python
df = df.drop("Memory", axis=1)
```

```python
df["Weight"] = df["Weight"].astype(float)
```

```python
df["GpuBrand"] = df.Gpu.str.split(" ").apply(lambda x: x[0])
df = df.drop("Gpu", axis=1)
```

```python
df = df.join(pd.get_dummies(df.OpSys))
df = df.drop("OpSys", axis=1)
```

```python
cpu_catagories = pd.get_dummies(df["CpuBrand"])
cpu_catagories.columns = [col + "_cpu" for col in cpu_catagories.columns]

df = df.join(cpu_catagories)
df = df.drop("CpuBrand", axis=1)
```

```python
gpu_catagories = pd.get_dummies(df["GpuBrand"])
gpu_catagories.columns = [col + "_gpu" for col in gpu_catagories.columns]

df = df.join(gpu_catagories)
df = df.drop("GpuBrand", axis=1)
```

```python
df
```

```python
df.columns
```

```python
df = df.join(pd.get_dummies(df.MemoryType))
df = df.drop("MemoryType", axis=1)
df
```

```python
import seaborn as sns

target_corrs = df.corr()["Price_euros"].apply(abs).sort_values()
```

```python
target_corrs
```

```python
selected_features =  target_corrs[-27:].index
```

```python
selected_features = list(selected_features)
```

```python
selected_features
```

```python
limited_df =df[selected_features]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X,y = limited_df.drop(["Price_euros"], axis=1), limited_df["Price_euros"]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15)

scalar = StandardScaler()

X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
forest = RandomForestRegressor()
forest.fit(X_train_scaled, y_train)
```

```python
forest.score(X_test_scaled, y_test)
```

```python
y_pred = forest.predict(X_train_scaled)
```

```python
X_new_scaled = scalar.transform([X_test.iloc[10]])
forest.predict(X_new_scaled)
```
